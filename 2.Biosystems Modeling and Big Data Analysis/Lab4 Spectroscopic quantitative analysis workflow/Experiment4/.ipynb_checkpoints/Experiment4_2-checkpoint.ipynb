课程大纲：

2、使用KFold函数对训练集样本进行交叉验证；   
3、构建PLS-LDA和PCA-LogReg定性分析模型，并根据交叉验证结果优化模型超参数；   
4、根据Overall accuracy和Confusion matrix指标分析并对比结果；   
5、结合SG Filter和Min-max scaling预处理方法，探索数据预处理对模型分类结果的影响。   

```python
import numpy as np
import pandas as pd
import random

from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.cross_decomposition import PLSRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from scipy.signal import savgol_filter

import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize 
```

```python
# 定义X,y划分函数
def X_y_split(df_raw):
    df_X = df_raw.iloc[:,0:df_raw.shape[1]-1]  # iloc 根据行号来索引  loc 根据行index来索引
    df_y = df_raw.iloc[:,df_raw.shape[1]-1]
    return df_X, df_y
```

```python
df_raw = pd.read_csv('Grapevine_5classes.csv',header=0,index_col=0)
df_train, df_test = train_test_split(df_raw, test_size=0.2, random_state=42,
                                     shuffle=True, stratify=df_raw['label'])
X_train, y_train = X_y_split(df_train)
X_test, y_test = X_y_split(df_test)
```

## 1. 预处理方法可视化

```python
# SG平滑处理
X_train_SG = savgol_filter(X_train, window_length=13, polyorder=2) # 人为设定用于平滑的多项式窗口大小和次数
X_test_SG = savgol_filter(X_test, window_length=13, polyorder=2)
X_train_SG = pd.DataFrame(X_train_SG, index=X_train.index, columns=X_train.columns)
X_test_SG = pd.DataFrame(X_test_SG, index=X_test.index, columns=X_test.columns)
df_train_SG = pd.concat([X_train_SG ,y_train], axis=1)
```

```python
# 查看SG平滑结果
figsize(6, 4)
px = np.linspace(1595.7, 2396.3, num=100)
py = X_train_SG.iloc[1,:]
plt.plot(px, py, color='red', label='SG processed')
plt.plot(px, X_train.iloc[1,:], color='blue', label='Raw data')
plt.grid(linestyle='--', which='major', color = 'lightgray')
plt.rcParams['figure.dpi'] = 300 
plt.legend()
```

```python
# 定义Min-max归一化方法
def scale_column(X_train, X_test):
    scaler = preprocessing.MinMaxScaler().fit(X_train)
    X_train_scaled = scaler.transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    return X_train_scaled, X_test_scaled
```

```python
# Min-max归一化原始数据
X_train_MS, X_test_MS = scale_column(X_train, X_test)
X_train_MS = pd.DataFrame(X_train_MS, index=X_train.index, columns=X_train.columns)
X_test_MS = pd.DataFrame(X_test_MS, index=X_test.index, columns=X_test.columns)
df_train_MS = pd.concat([X_train_MS, y_train], axis=1)
```

```python
# 查看Min-max归一化结果
figsize(6, 4)
px = np.linspace(1595.7, 2396.3, num=100)
colors = ['blue', 'lime', 'red', 'cyan', 'orange']
for item in range(X_train.shape[0]):
    py = X_train_MS.iloc[item,:]
    plt.plot(px, py, color=random.choice(colors))
plt.grid(linestyle='--', which='major', color='lightgray')
plt.rcParams['figure.dpi'] = 300 
```

```python
# SG平滑 + Min-max归一化
X_train_SG_MS, X_test_SG_MS= scale_column(X_train_SG, X_test_SG)
X_train_SG_MS = pd.DataFrame(X_train_SG_MS, index=X_train.index, columns=X_train.columns)
X_test_SG_MS = pd.DataFrame(X_test_SG_MS, index=X_test.index, columns=X_test.columns)
df_train_SG_MS = pd.concat([X_train_SG_MS,y_train], axis=1)
```

```python
# 查看SG平滑+Min-max归一化结果
figsize(6, 4)
px = np.linspace(1595.7, 2396.3, num=100)
colors = ['blue', 'lime', 'red', 'cyan', 'orange']
for item in range(X_train.shape[0]):
    py = X_train_SG_MS.iloc[item,:]
    plt.plot(px, py, color=random.choice(colors))
plt.grid(linestyle='--',which='major',color = 'lightgray')
plt.rcParams['figure.dpi'] = 300 
```

## 2. 生成交叉验证的数据索引

```python
# 生成交叉验证的子集index
kf = KFold(n_splits=5, shuffle=True, random_state=42)
lst_train_index = []
lst_valid_index = []
for train_index, valid_index in kf.split(df_train):
    lst_train_index.append(train_index)
    lst_valid_index.append(valid_index)
```

## 3. 处理SG平滑后的数据

```python
# PLS LDA optimization 
LV_list = []
for k in range(1,21):
    performance_list = []
        
    for train_index, valid_index in zip(lst_train_index, lst_valid_index):
        train_set = df_train_SG.iloc[train_index,:]
        valid_set = df_train_SG.iloc[valid_index,:]
        X_train_set, y_train_set = X_y_split(train_set)
        X_valid_set, y_valid_set = X_y_split(valid_set)

        pls = PLSRegression(n_components=k, scale=False)
        pls.fit(X_train_set,y_train_set)
        X_train_set_score = pls.transform(X_train_set, Y=None, copy=True)
        X_valid_set_score = pls.transform(X_valid_set, Y=None, copy=True)

        lda = LinearDiscriminantAnalysis()
        lda.fit(X_train_set_score, y_train_set)
        y_train_set_pred = lda.predict(X_train_set_score)
        y_valid_set_pred = lda.predict(X_valid_set_score)
        acc_train_set = accuracy_score(y_train_set, y_train_set_pred)
        acc_valid_set = accuracy_score(y_valid_set, y_valid_set_pred)
        performance_list.append(acc_valid_set)
    LV_list.append(np.mean(performance_list))
optim_LV = np.argmax(LV_list)+1
```

```python
# 测试PLS-LDA模型
pls_optim = PLSRegression(n_components=optim_LV, scale=False)
pls_optim.fit(X_train_SG, y_train)
X_train_score = pls_optim.transform(X_train_SG, Y=None, copy=True)
X_test_score = pls_optim.transform(X_test_SG, Y=None, copy=True)
lda_optim = LinearDiscriminantAnalysis()
lda_optim.fit(X_train_score, y_train)
y_train_pred = lda_optim.predict(X_train_score)
y_test_pred = lda_optim.predict(X_test_score)

acc_train = accuracy_score(y_train, y_train_pred)
acc_test = accuracy_score(y_test, y_test_pred)
print('training_accuracy:{:.4}, test_accuracy:{:.4}'.format(acc_train, acc_test))
print('test_confusion_matrix:\n',confusion_matrix(y_test, y_test_pred))
```

```python
result_table = pd.read_csv('result_table.csv',header=0,index_col=0)
result_SG_PLSLDA = {'Approaches': 'SG-PLS-LDA',
                  'Train accuracy': acc_train,
                  'Test accuracy': acc_test}
result_table = result_table.append(result_SG_PLSLDA, sort=False, ignore_index=True)
```
