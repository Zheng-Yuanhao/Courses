```python
import numpy as np
import pandas as pd
import random

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.cross_decomposition import PLSRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold

import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize 
```

## 1. 数据读取与画图

```python
df_raw = pd.read_csv('Grapevine_5classes.csv', header=0, index_col=0)
```

```python
df_raw
```

```python
# 画图查看数据
figsize(6, 4)
colors = ['blue', 'lime', 'red', 'cyan', 'orange']
px = np.linspace(1595.7, 2396.3, num=100)
for item in range(df_raw.shape[0]):
    py = df_raw.iloc[item,0:df_raw.shape[1]-1]
    # 对每一条光谱产生随机的颜色
    # plt.plot(px, py, color = random.choice(colors))
    # 对每一类的光谱产生不同的颜色
    plt.plot(px, py, color= colors[df_raw.iloc[item,df_raw.shape[1]-1]-1])
plt.grid(linestyle='--', which='major', color = 'lightgray')
plt.rcParams['figure.dpi'] = 300 
```

## 2. 均匀划分训练和测试数据

```python
df_train, df_test = train_test_split(df_raw, test_size=0.2, random_state=42,
                                     shuffle=True, stratify=df_raw['label']) # df_raw.iloc[:-1]
```

```python
# 检查总体数据是否满足8:2划分
print(df_train.shape, df_test.shape)
```

```python
# 检查每一类数据是否满足8:2划分
print(df_train[df_train['label']==1].shape, df_test[df_test['label']==1].shape)
```

## 3. 对训练集数据产生交叉验证的样本

```python
df_train.shape
```

```python
# kf = KFold(n_splits=5,  shuffle=True, random_state=42)
kf = KFold(n_splits=5, shuffle=False)
for train_index, valid_index in kf.split(df_train):
    print(train_index, valid_index)
```

```python
kf = KFold(n_splits=5, shuffle=False)
# kf = KFold(n_splits=5,  shuffle=True, random_state=42)
lst_train_index = []
lst_valid_index = []
for train_index, valid_index in kf.split(df_train):
    lst_train_index.append(train_index)
    lst_valid_index.append(valid_index)
```

```python
lst_train_index[0]
```

```python
lst_valid_index[0]
```

## 4. 定义自变量X与因变量y的数据划分

```python
# 定义X,y划分函数
def X_y_split(df_raw):
    df_X = df_raw.iloc[:,0:df_raw.shape[1]-1]  
    df_y = df_raw.iloc[:,df_raw.shape[1]-1]
    return df_X, df_y
```

```python
# iloc 根据行号来索引, loc 根据行index来索引
# df_raw.loc[1:,:]
# df_raw.iloc[1:,:]
```

```python
X_train, y_train = X_y_split(df_train)
X_test, y_test = X_y_split(df_test)
```

```python
print(train_index)
```

## 5. PLS-LDA模型运算

```python
# 根据交叉验证结果选择PLS超参数
LV_list = []
for k in range(1,21):
    performance_list = []
    for train_index, valid_index in zip(lst_train_index, lst_valid_index):
        train_set = df_train.iloc[train_index,:]
        valid_set = df_train.iloc[valid_index,:]
        X_train_set, y_train_set = X_y_split(train_set)
        X_valid_set, y_valid_set = X_y_split(valid_set)
        # 使用PLS模型，得到降维结果Score
        pls = PLSRegression(n_components=k, scale=False)
        # 模型拟合
        pls.fit(X_train_set,y_train_set)
        # 产生PLS降维后分数
        X_train_set_score = pls.transform(X_train_set, Y=None,copy=True)
        X_valid_set_score = pls.transform(X_valid_set, Y=None,copy=True)
        # 构建LDA分类器  
        lda = LinearDiscriminantAnalysis()
        # 模型拟合
        lda.fit(X_train_set_score, y_train_set)
        # 模型预测
        y_train_set_pred = lda.predict(X_train_set_score)
        y_valid_set_pred = lda.predict(X_valid_set_score)
        #计算准确率得分
        acc_train_set = accuracy_score(y_train_set, y_train_set_pred)
        acc_valid_set = accuracy_score(y_valid_set, y_valid_set_pred)
        performance_list.append(acc_valid_set)
    LV_list.append(np.mean(performance_list))
optim_LV = np.argmax(LV_list)+1
```

```python
# 产生五折交叉验证所产生的验证集效果
performance_list
```

```python
# 将平均后的结果输出，即每个超参数下对应的平均结果
LV_list
```

```python
optim_LV
```

```python
# 已知最优的超参数，测试PLS-LDA模型
pls_optim = PLSRegression(n_components=optim_LV,scale=False)
pls_optim.fit(X_train,y_train)
X_train_score = pls_optim.transform(X_train, Y=None, copy=True)
X_test_score = pls_optim.transform(X_test, Y=None, copy=True)

lda_optim = LinearDiscriminantAnalysis()
lda_optim.fit(X_train_score, y_train)
y_train_pred = lda_optim.predict(X_train_score)
y_test_pred = lda_optim.predict(X_test_score)

acc_train = accuracy_score(y_train, y_train_pred)
acc_test = accuracy_score(y_test, y_test_pred)
print('training_accuracy:{:.4}, test_accuracy:{:.4}'.format(acc_train, acc_test))
print('test_confusion_matrix:\n',confusion_matrix(y_test, y_test_pred))
```

```python
result_table = pd.DataFrame(columns=['Approaches','Train accuracy', 'Test accuracy'])
result_PLSLDA = {'Approaches': 'PLS-LDA',
                  'Train accuracy': acc_train,
                  'Test accuracy': acc_test}
result_table = result_table.append(result_PLSLDA, sort=False, ignore_index=True)
```

```python
result_table
```

## PCA-LogReg

```python
# 根据交叉验证结果选择PCA超参数
PC_list = []
for k in range(1,21):
    performance_list = []
    for train_index, valid_index in zip(lst_train_index, lst_valid_index):
        train_set = df_train.iloc[train_index,:]
        valid_set = df_train.iloc[valid_index,:]
        X_train_set, y_train_set = X_y_split(train_set)
        X_valid_set, y_valid_set = X_y_split(valid_set)

        # 使用PCA模型，得到降维结果Score
        pca = PCA(n_components=k)
        pca.fit(X_train_set)
        X_train_set_score = pca.transform(X_train_set)
        X_valid_set_score = pca.transform(X_valid_set)

        # 构建LogReg分类器  
        LogR = LogisticRegression() #选择逻辑回归的优化器
        LogR.fit(X_train_set_score, y_train_set)
        y_train_set_pred = LogR.predict(X_train_set_score)
        y_valid_set_pred = LogR.predict(X_valid_set_score)
        acc_train_set = accuracy_score(y_train_set, y_train_set_pred)
        acc_valid_set = accuracy_score(y_valid_set, y_valid_set_pred)
        performance_list.append(acc_valid_set)
    PC_list.append(np.mean(performance_list))
optim_PC = np.argmax(PC_list)+1
```

```python
optim_PC
```

```python
# 测试PCA-LogReg模型
pca_optim = PCA(n_components=optim_PC)
pca_optim.fit(X_train)
X_train_score = pca_optim.transform(X_train)
X_test_score = pca_optim.transform(X_test)

LogR_optim = LogisticRegression()
LogR_optim.fit(X_train_score, y_train)
y_train_pred = LogR_optim.predict(X_train_score)
y_test_pred = LogR_optim.predict(X_test_score)

acc_train = accuracy_score(y_train, y_train_pred)
acc_test = accuracy_score(y_test, y_test_pred)
print('training_accuracy:{:.4}, test_accuracy:{:.4}'.format(acc_train, acc_test))
print('test_confusion_matrix:\n',confusion_matrix(y_test, y_test_pred))
```

```python
result_PCALogR = {'Approaches': 'PCA-LogR',
                  'Train accuracy': acc_train,
                  'Test accuracy': acc_test}
result_table = result_table.append(result_PCALogR, sort=False, ignore_index=True)
```
