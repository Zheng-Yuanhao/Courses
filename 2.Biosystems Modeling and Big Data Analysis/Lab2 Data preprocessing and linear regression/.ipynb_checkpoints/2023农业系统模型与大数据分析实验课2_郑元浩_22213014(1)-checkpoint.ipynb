## 气象数据部分

```python
# 查看气象数据前5行 
df_clim.head(5)
```

```python
# 提取月份作为一列，用于后续的筛选与聚合
df_clim['new_date'] = pd.to_datetime( df_clim['date'] )
df_clim['month'] = df_clim['new_date'].dt.month
df_clim = df_clim.drop( 'new_date', axis = 1 )
df_clim.head(5)
```

```python
# 筛选月份，美国玉米5月播种，9月成熟
df_clim = df_clim[ df_clim['month'].isin([5,6,7,8,9]) ]
df_clim
```

```python
# 气象数据概览
df_clim.describe()
```

```python
# 查看各气象原始值的分布，例如绘制直方图
t = 0
for i in ['maxt', 'mint', 'avgt', 'pcpn']:
    t+=1
    plt.figure(t)
    plt.hist(df_clim[i])
    plt.xlabel(i)
```

```python
# 查看各月降雨量分布
df_clim.boxplot('pcpn', by='month')
```

```python
# 计算日值气象指标 （GDD, KDD）
df_clim['Tmax'] = df_clim['maxt']
for i,T in df_clim['Tmax'].iteritems():
    if T>=29:
        df_clim['Tmax'][i] = 29
    elif T<=10:
        df_clim['Tmax'][i] = 10

df_clim['Tmin'] = df_clim['mint']
for i,T in df_clim['Tmin'].iteritems():
    if T>=29:
        df_clim['Tmin'][i] = 29
    elif T<=10:
        df_clim['Tmin'][i] = 10

df_clim['GDD'] = (df_clim['Tmax']+df_clim['Tmin'])/2 - 10

df_clim['KDD'] = df_clim['maxt']-29
for i, kdd in df_clim['KDD'].iteritems():
    if kdd<0:
        df_clim['KDD'][i] = 0
        
df_clim
```

```python
# 查看气象指标的数据概览
t = 0
for i in ['Tmax','Tmin','GDD', 'KDD']:
    t+=1
    plt.figure(t)
    plt.hist(df_clim[i])
    plt.xlabel(i)

df_clim[['Tmax','Tmin','GDD', 'KDD']].describe()
```

```python
# 计算年累积值以查看年际差异
plt.plot( df_clim.groupby(df_clim['Year'])[['KDD', 'GDD','pcpn']].sum() )
plt.legend( ['KDD', 'GDD','pcpn'] )
plt.xlabel( 'Year' )
plt.ylabel( 'SumData' )
```

```python
# 查看气象数据在每年的分布情况
plt.plot( df_clim.groupby(df_clim['Year'])[['maxt', 'mint','avgt','pcpn']].sum() )
plt.legend(['maxt', 'mint','avgt','pcpn'] )
plt.xlabel( 'Year' )
plt.ylabel( 'SumData' )
```

```python
# 将日值聚合成月值: 根据fips，Year和Mon，对KDD,GDD,pcpn进行分组求和，即每个FIPS在每一年在每个月份每个气象指标有一个sum值
MergeData = df_clim.groupby(["fips","Year","month"])[['pcpn','KDD','GDD']].sum()
MergeData
```

```python
# 行列的融合拆分操作1
split_rows = MergeData.reset_index()
FYM = MergeData.pivot_table(index=["fips", "Year"], columns="month", values=["KDD", "GDD", "pcpn"])
FYM
```

```python
#产量数据
df_yield
```

```python
# 将气象数据与单产数据合并，形成样本集
df = FYM.merge(df_yield, on=["fips", "Year"])
df
```

```python
# 查看合并前后数据维度
print("FYM 维度:", FYM.shape)
print("df_yield 维度:", df_yield.shape)
print("合并后的数据集 df 维度:", df.shape)
```

```python
# 将建立好的样本集导出成CSV文件

new_columns = ['fips', 'Year', 'GDD5', 'GDD6', 'GDD7', 'GDD8', 'GDD9', 'KDD5', 'KDD6', 'KDD7', 'KDD8', 'KDD9', 'pcpn5', 'pcpn6', 'pcpn7', 'pcpn8', 'pcpn9', 'county', 'state', 'Ton_ha', 'state_fips']
df.columns = new_columns
df.to_csv('MergeDataSet_郑元浩_22213014.csv', index=False)

df_show =  pd.read_csv('MergeDataSet_郑元浩_22213014.csv',header=0,index_col=None)
df_show
```

# 2. 线性回归

目的：了解并熟悉线性回归的python实现方式并理解线性模型的统计分析结果：R方，调节R方，参数拟合值，t检验，F检验，p-value等

读取上一步工作生成的样本集，建立玉米单产对各月份GDD,KDD和PRCP的多元线性回归模型，回答下述问题：
1. 有多少产量变化可以由这些气象变量解释？（模型R方值）
2. 各变量对玉米单产影响最大的月份分别是？在对应月份玉米对该气象因子的敏感性为？

提示：公式为'Ton_ha~GDD5+GDD6+GDD7+GDD8+GDD9+KDD5+KDD6+KDD7+KDD8+KDD9+pcpn5+pcpn6+pcpn7+pcpn8+pcpn9+b' (b为截距项)

### 方式一 使用statsmodels

```python
# 引入模块
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# 读取数据
df = pd.read_csv('MergeDataSet_郑元浩_22213014.csv')
# 定义模型输入输出
x = df[['GDD5','GDD6','GDD7','GDD8','GDD9','KDD5','KDD6','KDD7','KDD8','KDD9','pcpn5','pcpn6','pcpn7','pcpn8','pcpn9']]
y = df['Ton_ha']

df
```

```python
# 构建最小二乘法线性模型

x = sm.add_constant(x,prepend=True)
model = sm.OLS(y, x)
results = model.fit()

predictions = results.predict(x)
results.summary()
```

```python
# 绘制 y真值 与 y预测值 的散点图
plt.scatter(y,predictions)
```

```python
# 构建最小二乘法线性模型
formula = 'Ton_ha~GDD5+GDD6+GDD7+GDD8+GDD9+KDD5+KDD6+KDD7+KDD8+KDD9+pcpn5+pcpn6+pcpn7+pcpn8+pcpn9'
model2 = smf.ols(formula=formula, data=df)
results2 = model2.fit()
print(results2.summary())
# 绘制 y真值 与 y预测值 的散点图
predictions2 = results2.predict(x)
plt.scatter(y,predictions2)
```

## 方式二 使用sklearn

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 读取数据
df = pd.read_csv('MergeDataSet_郑元浩_22213014.csv')
x = df[['GDD5', 'GDD6', 'GDD7', 'GDD8', 'GDD9', 'KDD5', 'KDD6', 'KDD7', 'KDD8', 'KDD9', 'pcpn5', 'pcpn6', 'pcpn7', 'pcpn8', 'pcpn9']]
y = df['Ton_ha']

# 划分训练集和预测集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# 定义模型
model = LinearRegression()

# 拟合模型
model.fit(x_train, y_train)
 
# 截距  回归系数  计算R方
print("截距：",model.intercept_)
print("线性模型回归系数：",model.intercept_)
print("R方R2_train：",model.score(x_train,y_train))
print("R方R2_test：",model.score(x_test,y_test))

y_pred = model.predict(x_test)
plt.scatter(y_test,y_pred)
```
